```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Read flight data
flight = pd.read_csv("flight delay.csv")

# Subset for Cincinnati and Chicago
cvg = flight[flight["Origin"] == "CVG"]["Delay"]
ord = flight[flight["Origin"] == "ORD"]["Delay"]

# Traditional bootstrap for Chicago flights
n_samples = 10000
bootstrap_means = np.array([ord.sample(len(ord), replace=True).mean() for _ in range(n_samples)])

# Plot histogram
sns.histplot(bootstrap_means, binwidth=2, kde=True)
plt.axvline(ord.mean(), color="green", linestyle="--", linewidth=1)
plt.axvline(bootstrap_means.mean(), color="red", linestyle="--", linewidth=1)
plt.show()
```

```python
# Bootstrap percentile interval
def percentile_interval(bootstrap_dist, loc=0.9):
    lower_limit = np.percentile(bootstrap_dist, (1 - loc) * 50)
    upper_limit = np.percentile(bootstrap_dist, 100 - (1 - loc) * 50)
    return lower_limit, upper_limit
```

```python
print(percentile_interval(bootstrap_means, loc=0.9))
```

```python
# Kernel density estimate
sns.histplot(cvg, binwidth=50, kde=True, kde_kws={"bw_adjust": cvg.std() / np.sqrt(len(cvg))})
plt.show()
```

```python
# Traditional bootstrap for Cincinnati
def first_quartile(data):
    return np.percentile(data, 25)

bootstrap_q1 = np.array([cvg.sample(len(cvg), replace=True).pipe(first_quartile) for _ in range(n_samples)])

# Plot histogram
sns.histplot(bootstrap_q1, binwidth=2, kde=True)
plt.axvline(first_quartile(cvg), color="red", linestyle="--", linewidth=1.5)
plt.axvline(bootstrap_q1.mean(), color="limegreen", linestyle="--", linewidth=1.5)
plt.show()
```

```python
# Smoothed bootstrap for Cincinnati
smoothed_bootstrap_q1 = []
for _ in range(n_samples):
    bootstrap_sample = cvg.sample(len(cvg), replace=True)
    random_noise = np.random.normal(loc=0, scale=cvg.std() / np.sqrt(len(cvg)), size=len(cvg))
    smoothed_bootstrap_q1.append(first_quartile(bootstrap_sample + random_noise))
smoothed_bootstrap_q1 = np.array(smoothed_bootstrap_q1)

# Plot histogram
sns.histplot(smoothed_bootstrap_q1, binwidth=1, kde=True)
plt.axvline(first_quartile(cvg), color="red", linestyle="--", linewidth=1.5)
plt.axvline(smoothed_bootstrap_q1.mean(), color="limegreen", linestyle="--", linewidth=1.5)
plt.show()
```

```python
import cartopy.crs as ccrs
import ipywidgets as widgets
from IPython.display import display
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
def plot_co2(time_index):
    # Create a new figure with appropriate size
    fig = plt.figure(figsize=(12,6))
    
    # Create an axis with the PlateCarree projection
    ax = plt.axes(projection=ccrs.PlateCarree())
    
    # Set the extent of the plot to cover the entire globe
    ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())

    # Add coastlines to the plot
    ax.coastlines()

    # Plot the CO2 data with appropriate latitude and longitude coordinates
    img = ax.pcolormesh(carbon_test.lon, carbon_test.lat, carbon_test.xco2[time_index,:,:], 
                        cmap='viridis', transform=ccrs.PlateCarree())

    # Add gridlines and labels for latitude and longitude
    gl = ax.gridlines(draw_labels=True, linestyle='--', alpha=0.5)
    gl.top_labels = False
    gl.right_labels = False
    gl.xlines = True
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER

    # Add colorbar and labels
    plt.colorbar(img, label='CO2 levels', 
                 orientation='vertical',location = 'left', pad=0.05)
    plt.title(f'Time index: {time_index}')
    
    plt.show()

slider = widgets.IntSlider(min=0, max=len(carbon_test.xco2)-1, 
                           step=1, value=0, description='Time index')
widgets.interact(plot_co2, time_index=slider)
```
